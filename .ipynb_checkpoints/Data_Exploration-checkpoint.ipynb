### Analysis

A ```count``` of each of the following 6 features are selected for visualization considering the sentence structure and basic statistics of an essay.

1. Domain Words: Does the word belong to the domain of the essay?
2. Word to sentence ratio : Are the sentences short and concise or long and convoluted?
3. Average Word Length: Are they using a sophisticated vocabulary?
4. Noun Count: Number of nouns used?
5. Long word count: Number of words with length > 6?
6. Spelling Mistakes: How often is the author making spelling mistakes?

These six key features should allow us to gain some insight into what is happening in our data. to calculate this, the **TextBlob, NLTK, Textmining etc** packages were used. Please refer ```Libraries.txt``` for a complete list of libraries used in this project.

### Visualization

The plots below will allow you to view the underlying trends within the data. The plots below visualize how the 6 features above impact the grade of an essay.

As expected, we were able to see general trends between the sentence structure of an essay and the grade it received. What was unexpected was the fact that the faults of an essay did not carry much weight. Noteably, we see that it is not uncommon to see essays with a large number of spelling mistakes receive a high grade. One could make the assumption that techonology's (Word's Spell Check) influence on writing has made it less imperitive for students to spell correctly. It seems as though the message carries more weight than specific errors in the essay.

# Free form visualization :
## Combines the word2vec model feature vectors and the heuristic feature vectors

Here, we will use Principal Component Analysis (PCA) for visualizing good essays (grade > 8) and poor essay (grade < 4) to see if any information can be obtained from the resultant graphs

Extracting 100 good essays and 100 poor essays for performing PCA:

````python
X_all['target'] = y_all
goodessay = pd.DataFrame()
pooressay = pd.DataFrame()

for index,rows in X_all.iterrows():
    if len(goodessay) == 100 and len(pooressay) == 100:
        break
    elif rows['target'] > 8 and len(goodessay) <= 100:
        goodessay = goodessay.append(rows)
    elif rows['target'] < 4 and len(pooressay) <= 100:
        pooressay = pooressay.append(rows)
print(len(pooressay))      
````

**Word2vec model already saved during the learning phase will be used for generating the feature vectors** 
Created a function ```essayvectors()``` that returns essay vectors for a given essay set. 

````python
from gensim.models import Word2Vec
essaygood = goodessay['essay']
essaypoor = pooressay['essay']

#Initializing variables for generating the word2vec model.
num_features = 300 
min_word_count = 40
num_workers = 4
context = 10
downsampling = 1e-3

#Function to generate essay vectors.
def essayvectors(essaysv):
        
    sentences = [] 
    #Extract sentences that will be used for generating the word vectors.
    for essay in essaysv:
        sentences += helper.essay_to_sentences(unicode(essay,"utf-8",errors='ignore'), remove_stopwords = True)
    
    #Using saved word2vec model during the learning phase.
    model = word2vec.Word2Vec(sentences, workers=num_workers, size=num_features, min_count = min_word_count, window = context, sample = downsampling)

    clean_train_essays = []
    
    #Generate the essay vectors
    for essay_v in essaysv:
        clean_train_essays.append( helper.essay_to_wordlist( essay_v, remove_stopwords=True ))
    essayvecs = helper.getAvgFeatureVecs( clean_train_essays, model, num_features )
      
    return essayvecs
````

**Performing the PCA**

````python
from sklearn.decomposition import PCA

vecs2 = essayvectors(essaypoor)
vecs1 = essayvectors(essaygood)

#Remove "nan" values if occured.
p = vecs2[~np.isnan(vecs2).any(axis=1)]

#For good essays.
pca1 = PCA(whiten=True) #Initialize PCA
pca1.fit(vecs1) #Fit the good essay vectors.
pcat1 = pca1.transform(vecs1) #Transform

#For poor essays
pca2 = PCA(whiten=True) #Initialize PCA
pca2.fit(p) #Fit the poor essay vectors
pcat2 = pca2.transform(p) #Transform
#Plotting the PCA transformed data.

#For Good essays
ys = [i[1] for i in pcat1] #PCA Component y, NOTE: This can be accomplished by using pca.components_[1]
xs = [i[0] for i in pcat1] #PCA Component x  NOTE: This can be accomplished by using pca.components_[0]

#For poor essays
ys2 = [i[1] for i in pcat2]
xs2 = [i[0] for i in pcat2]

fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(15, 5))
ax1.scatter(xs,ys,color='b')
ax1.set_ylabel('PCA component y')
ax1.set_xlabel('PCA component x')
ax1.set_title('PCA For Good essays (score > 8)')

ax2.scatter(xs2,ys2,color='r')
ax2.set_ylabel('PCA component y')
ax2.set_xlabel('PCA component x')
ax2.set_title('PCA For Poor essays (score < 4)')

plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlabel('PCA component x')
ax1.set_ylabel('PCA component y')
ax1.set_title('PCA for good and poor essays')
ax1.scatter(xs,ys, label='Good essay (grade > 8)')
ax1.scatter(xs2,ys2, label = 'Poor essay (grade <4)')
plt.legend()
plt.show()
````
